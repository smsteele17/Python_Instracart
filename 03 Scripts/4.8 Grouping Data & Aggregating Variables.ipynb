# 4.8 Grouping Data & Aggregating Variables

```python
#import libraries
import pandas as pd
import numpy as np
import os
```

```python
#set path
path = r'/Users/shelbysteele/Documents/Instacart Basket Anaysis-1022'
```

```python
#import data
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_derived.pkl'))
```

```python
#creating a subset of the million rows
df = ords_prods_merge[:1000000]
```

```python
#checking shape
df.shape
```

```python
#checking columns
df.head()
```

```python
#grouping subset by product_name
df.groupby('product_name')
```

```python
#grouping and aggegating using mean
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#using another function to find mean w/o agg()
df.groupby('department_id')['order_number'].mean()
```

```python
#finding multiple stats at the same time
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#grouping 
df.groupby('user_id')
```

```python
#using transform to gorup
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking columns to see 'max_order'"
ords_prods_merge.head()
```

```python
#creating flags for loyalty
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking output of flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
#viewing only 3 columns using head()
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
#drop unneeded columns
ords_prods_merge.drop(columns = {'new_customer'}, inplace = True)
```
