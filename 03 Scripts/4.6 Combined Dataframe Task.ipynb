```python
#5: Import product data
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'))
```

```python
#change data type for df_prods
df_prods['aisle_id'] = df_prods['aisle_id'].astype('int8')
df_prods['department_id'] = df_prods['department_id'].astype('int8')
df_prods['prices'] = df_prods['prices'].astype('float64')
df_prods['product_id'] = df_prods['product_id'].astype('int32')
```

```python
df_prods.info()
```

```python
#6: attempt to merge df_prods to df_ords_prods combinded data
df_ords_prods_merged = df_ords_prods.merge(df_prods, on = 'product_id', indicator = True)
```

```python
#remove unneeded columns
df_ords_prods_merged.drop(columns = {'Unnamed: 0.1', 'Unnamed: 0'}, inplace = True)
```
