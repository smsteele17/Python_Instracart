```python
#import Libraires
import pandas as pd
import numpy as np
import os
```

```python
#set path
path = r'/Users/shelbysteele/Documents/Instacart Basket Anaysis-1022'
```

```python
#import data
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))
```

```python
#creating a subset of the million rows
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
#creating a condition for prices
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#checking frequency 
df['price_range'].value_counts(dropna = False)
```

```python
#checking max value matches funciton output
df['prices'].max()
```

```python
#creating a new function for loc()
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#checking frequency of new function
df['price_range_loc'].value_counts(dropna = False)
```

```python
#using loc() on entire dataframe and not just a subset
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#checking frequency of new function on entire dataframe
df_ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

```python
#checking frequency on new column for new function
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
#creating a loop for orders_day_of_week 
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#result list
result
```

```python
#attaching result list to dataframe
df_ords_prods_merged['busiest_day'] = result 
```

```python
#checking frequency of new column
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# 4.7 Task

```python
#2: Create a new column for "Busiest Days of the week"
result2 = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value in [0,1]:
    result2.append("Busiest days")
  elif value in [3,4]:
    result2.append("Slowest days")
  else:
    result2.append("Regularly busy")
```

```python
#combine result2 with dataframe
df_ords_prods_merged['Busiest_days'] = result2
```

```python
#3: checking frequency of new column
df_ords_prods_merged['Busiest_days'].value_counts(dropna = False)
```

#### There are more busy and regular days compared to slow days. 

```python
#4: What are the busiest hours of the day? 
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#grouping hours of day into time periods
result3 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    result3.append("Most Orders")
  elif value in [23,6,0,1,5,2,4,3]:
    result3.append("Fewest Orders")
  else:
    result3.append("Average Orders")
```

```python
#combine result3 with dataframe
df_ords_prods_merged['busiest_period_of_day'] = result3
```

```python
#5: checking frequency of new column
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
